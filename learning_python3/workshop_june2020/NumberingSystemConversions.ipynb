### Binary <-> Decimal

```python
my_var = int("1010", base=2)
my_var
```

```python
print(f'The integer is = {my_var}')
print(f'Binary representation = {my_var:b}')
```

```python
bin(my_var)
```

### Octal <-> Decimal

```python
my_var = int("11", base=8)
my_var
```

```python
print(f'The integer is = {my_var}')
print(f'Octal representation = {my_var:o}')
```

```python
oct(my_var)
```

### Hexadecimal <-> Decimal

```python
my_var = int("F", base=16)
my_var
```

```python
print(f'The integer is = {my_var}')
print(f'Hexadecimal representation = {my_var:x}')
print(f'Hexadecimal representation = {my_var:X}')
```

```python
hex(my_var)
```
